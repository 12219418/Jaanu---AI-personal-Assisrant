```python
import speech_recognition as sr
import pyttsx3
import datetime
import wikipedia
import pyjokes
import os
import sys
import sounddevice as sd
import numpy as np
import webbrowser
import pywhatkit  
# Initialize Text-to-Speech
engine = pyttsx3.init()
engine.setProperty('rate', 170)
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[1].id)  

def talk(text):
    """Speak and print text."""
    print(f"üéôÔ∏è Jaanu: {text}")
    engine.say(text)
    engine.runAndWait()
# Take Voice Command

def take_command():
    recognizer = sr.Recognizer()
    samplerate = 16000
    duration = 5
    talk("Please say something üéß")
    print("üéß Listening...")

    try:
        # Record user's speech
        audio = sd.rec(int(samplerate * duration), samplerate=samplerate, channels=1, dtype='int16')
        sd.wait()
        audio_data = sr.AudioData(audio.tobytes(), samplerate, 2)

        # Recognize speech
        command = recognizer.recognize_google(audio_data)
        command = command.lower()
        print(f"üó£Ô∏è You said: {command}")  
        talk(f"You said: {command}")
        return command

    except sr.UnknownValueError:
        talk("Sorry, I didn‚Äôt catch that.")
        return ""
    except sr.RequestError:
        talk("Network issue with Google service.")
        return ""
    except Exception as e:
        print("‚ö†Ô∏è Error:", e)
        talk("Something went wrong while listening.")
        return ""

# Play Telugu Songs 
def play_telugu_song(song_name):
    """Play Telugu song directly from YouTube search."""
    try:
        search_query = f"{song_name} Telugu song"
        talk(f"Playing {song_name} from Telugu songs üé∂")
        pywhatkit.playonyt(search_query)
    except Exception as e:
        print("‚ö†Ô∏è Error playing song:", e)
        talk("Sorry, I couldn‚Äôt play that Telugu song.")

# Wikipedia Search 
def search_wikipedia(query):
    """Improved Wikipedia search with better error handling."""
    try:
        query = query.strip()
        if not query:
            talk("Please tell me what you want to know.")
            return
        
        wikipedia.set_lang("en")
        talk(f"Searching Wikipedia for {query} üß†")

        results = wikipedia.search(query)
        if not results:
            talk("No results found on Wikipedia.")
            return

        summary = wikipedia.summary(results[0], sentences=2, auto_suggest=False)
        talk(summary)

    except wikipedia.DisambiguationError as e:
        talk(f"There are multiple meanings for {query}. Please be more specific.")
        print("üîç Options:", e.options[:5])
    except wikipedia.PageError:
        talk("I couldn‚Äôt find a Wikipedia page for that topic.")
    except Exception as e:
        print("‚ö†Ô∏è Wikipedia Error:", e)
        talk("Something went wrong while fetching information.")

# Main Jaanu Logic

def run_jaanu():
    command = take_command()
    if not command:
        return

    # Play Telugu Song
    if "play" in command:
        song = command.replace("play", "").strip()
        play_telugu_song(song)

    #  Google Search
    elif "search" in command:
        query = command.replace("search", "").strip()
        if query:
            talk(f"Searching Google for {query} üåê")
            webbrowser.open(f"https://www.google.com/search?q={query}")
        else:
            talk("Please tell me what to search for.")

    #  Time
    elif "time" in command:
        time = datetime.datetime.now().strftime('%I:%M %p')
        talk(f"It‚Äôs {time} ‚è∞")

    #  About Sai
    elif "sai" in command:
        talk("Sai is a passionate programmer and tech enthusiast who loves creating amazing Python projects üíª")
        talk("Would you like me to open Sai's Instagram page?")
        reply = take_command()
        if "yes" in reply or "sure" in reply:
            talk("Opening Sai's Instagram üî•")
            webbrowser.open("https://www.instagram.com/sai/")
        else:
            talk("Okay, no problem üëç")

    #  Wikipedia
    elif "who is" in command or "what is" in command or "tell me about" in command:
        topic = (
            command.replace("who is", "")
            .replace("what is", "")
            .replace("tell me about", "")
            .strip()
        )
        search_wikipedia(topic)

    #  Open Chrome
    elif "open chrome" in command:
        chrome_path = "C:\\Program Files\\Google\\Chrome\\Application\\chrome.exe"
        if os.path.exists(chrome_path):
            talk("Opening Chrome üöÄ")
            os.startfile(chrome_path)
        else:
            talk("Chrome not found üò¨")

    # üëã Exit
    elif "exit" in command or "stop" in command or "quit" in command:
        talk("Okay, see you later üëã")
        sys.exit()

    else:
        talk("I heard you, but I don‚Äôt understand that yet üòÖ")

# ---------------------------
# Start Assistant
# ---------------------------
talk("Hey! I'm Jaanu ‚Äì YOUR Personal Assistant üé∂üí´")
while True:
    run_jaanu()
```
